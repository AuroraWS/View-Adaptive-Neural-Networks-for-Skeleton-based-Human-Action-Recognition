```python
import os

def get_filenames_without_extension(directory):
    # 检查目录是否存在
    if not os.path.exists(directory):
        print(f"The directory {directory} does not exist.")
        return []
    else:
      print(f'The directory {directory} does exist.')

    filenames = []
    # 遍历目录中的所有文件
    for filename in os.listdir(directory):
        name, _ = os.path.splitext(filename)
        filenames.append(name)
        # print(f'{name} has been processed')

    return filenames

def write_filenames_to_file(filenames, output_file):
    with open(output_file, 'w') as f:
        for name in filenames:
            f.write(name + '\n')

# 使用示例
directory_path = '/content/drive/Othercomputers/我的笔记本电脑/nturgb+d_skeletons'  # 替换为你的目录路径
output_file = '/content/drive/Othercomputers/我的笔记本电脑/View-Adaptive-Neural-Networks-for-Skeleton-based-Human-Action-Recognition/data/ntu/statistics/skes_available_name.txt'

filenames = get_filenames_without_extension(directory_path)
write_filenames_to_file(filenames, output_file)

print(f"Filenames have been written to {output_file}")

```

```python
def read_filenames_from_file(input_file):
    with open(input_file, 'r') as f:
        filenames = f.read().splitlines()
    return filenames

def sort_filenames(filenames):
    def extract_keys(name):
        # 提取 S、C、P、R、A 后面的数字
        s = int(name[1:4])
        c = int(name[5:8])
        p = int(name[9:12])
        r = int(name[13:16])
        a = int(name[17:20])
        return (s, c, p, r, a)

    # 按照提取的键进行排序
    return sorted(filenames, key=extract_keys)

# 使用示例
input_file = '/content/drive/Othercomputers/我的笔记本电脑/View-Adaptive-Neural-Networks-for-Skeleton-based-Human-Action-Recognition/data/ntu/statistics/skes_available_name.txt'

filenames = read_filenames_from_file(input_file)
sorted_filenames = sort_filenames(filenames)
write_filenames_to_file(sorted_filenames, input_file)

print(f"Sorted filenames have been written to {input_file}")
```

```python
def extract_numbers_from_filename(filename):
    # 提取 S、C、P、R、A 后面的数字
    s = int(filename[1:4])
    c = int(filename[5:8])
    p = int(filename[9:12])
    r = int(filename[13:16])
    a = int(filename[17:20])
    return s, c, p, r, a

def clear_files(*files):
  for file in files:
      open(file, 'w').close()

def write_numbers_to_files(filenames, setup_file, camera_file, performer_file, replication_file, label_file):
  # 清空文件内容
  clear_files(setup_file, camera_file, performer_file, replication_file, label_file)
  with open(setup_file, 'w') as sf, \
         open(camera_file, 'w') as cf, \
         open(performer_file, 'w') as pf, \
         open(replication_file, 'w') as rf, \
         open(label_file, 'w') as lf:

    for filename in filenames:
      # s, c, p, r, a = extract_numbers_from_filename(filename)
      result = extract_numbers_from_filename(filename)
      e=0
      if result:  # Check if extraction was successful
        s, c, p, r, a = result
      else:
        print(e)
      sf.write(f"{s}\n")
      cf.write(f"{c}\n")
      pf.write(f"{p}\n")
      rf.write(f"{r}\n")
      lf.write(f"{a}\n")

base_path = '/content/drive/Othercomputers/我的笔记本电脑/View-Adaptive-Neural-Networks-for-Skeleton-based-Human-Action-Recognition/data/ntu/statistics'
input_file = os.path.join(base_path,'skes_available_name.txt')
setup_file = os.path.join(base_path,'setup.txt')
camera_file = os.path.join(base_path,'camera.txt')
performer_file = os.path.join(base_path,'performer.txt')
replication_file = os.path.join(base_path,'replication.txt')
label_file = os.path.join(base_path,'label.txt')

# 读取文件名
with open(input_file, 'r') as f:
    skes_names = f.read().splitlines()
print(skes_names[0],extract_numbers_from_filename(skes_names[0]))

# 写入各个文件
write_numbers_to_files(skes_names, setup_file, camera_file, performer_file, replication_file, label_file)

print("Numbers have been extracted and written to the respective files.")
```
